```python
import matplotlib.pyplot as plt

# Expertise areas and corresponding values
fields = ["Data Analysis", "Software Development", "Entrepreneurship", "AI & Computer Vision"]
values = [8, 8, 7, 8]  # Expertise level (out of 10)

# Create radar chart
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True), facecolor='none')

# Convert data for plotting
angles = [n / float(len(fields)) * 2 * 3.1416 for n in range(len(fields))]
values += values[:1]
angles += angles[:1]

# Plot data
ax.fill(angles, values, 'skyblue', alpha=0.4)  # Softer color
ax.plot(angles, values, 'steelblue', linewidth=2)  # Darker outline

# Add values with "/10" format
for i, angle in enumerate(angles[:-1]):
    ax.text(angle, values[i] + 0.3,values[i],
            ha='center', va='center', color='black', fontsize=12,
            fontweight='bold')

# Set labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(fields, fontsize=12, color='dimgray', fontweight='medium')
ax.set_yticks([2, 4, 6, 8, 10])
ax.set_yticklabels(["2", "4", "6", "8", "10"])
ax.set_ylim(0, 10)
ax.spines['polar'].set_visible(False)

# Title
plt.title("Expertise Field Chart", fontsize=16, fontweight='bold',color='White', pad=60)

# Add a scale annotation
plt.text(0, -1.5, "Scale: 0 to 10", fontsize=12, color='dimgray', ha='center')

# Grid
ax.grid(True, color='lightgray', linestyle='--')

plt.show()
```
